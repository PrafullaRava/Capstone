## Methodology <a name="methodology"></a>

In this section, I will describe the data analysis and how I used the data to yield the results.

Starting out, I scraped data from Wikipedia to create a dataframe with the city districts of Delhi: https://en.wikipedia.org/wiki/List_of_districts_of_Delhi. For this, I used the pandas read function. I had to clean the resulting data frame in terms of unnecessary information or data that could not be handled in a data frame, such as picture data of the coat of arms of each district. The result is a nice data frame:

Delhi sub.PNG

Then, I enabled geopy functions by installing the conda-forge geopy package. I used the nominatim function to add geospatial data to the data frame, that is the latitude and the longitude seen on the right side of the following table.

delhi district3.PNG

Using the folium package and my data frame, I then created a map with the nine city districs on it.

delhi district4.PNG

Now, foursquare data comes into play.

I retrieved the foursquare data for all venues on foursquare with a distance of less than 3000 meters from each center of each city district, as indicated as blue dots in the map above. The result was a list of 757 venues all over Delhi. Out of these 757 venues, 184 where restaurants. These 184 restaurants come from 35 unique restaurant categories, such as Indian, Chinese or Asian.

I plotted a bar chart with the frequency of the 10 most frequently occuring restaurants in the whole city, using seaborn/matplotlib packages. We can see that Indian, Fast food and Chinese restaurants are the most frequently occuring restaurants in Delhi, which seems pretty reasonable, as these cuisines being excellent and highly appreciated by large parts of the population - think about biryani,noodles or kebap!

delhi district5.PNG

To find clusters of restaurant types in the different city districts, I first transformed the data frame with the restaurant venues, associated to city districts, by one-hot encoding (0/1), as seen in the picture below.

delhi district6.PNG

Next, I used grouping to show the frequency of each category of restaurants in each city district.

delhi district7.PNG

I used this information to create a data frame in which you can see the most common restaurant venue types for each city district.

delhi district8.PNG

Now, with all this data, I could finally run an unsupervised machine learning algorithm, more specifically, a k-means clustering algorithm from the scikit-learn package. One could use the ellbow method to systematically define the k value, but I simply chose k to be 5, having been inspired by one of the coursera courses to do so.
